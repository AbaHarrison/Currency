```python
def convert_to_ghs(row):
    if row['Currency'] == 'USD':
        return row['Amount']*11.63
    
    elif row['Currency'] == 'EUR':
        return row['Amount']*12.22
    
    elif row['Currency'] == 'CHF':
        return row['Amount']*12.66
    
    elif row['Currency'] == 'SEK':
        return row['Amount']*1.05
    
    elif row['Currency'] == 'RON':
        return row['Amount']*2.45
    
    elif row['Currency'] == 'PLN':
        return row['Amount']*2.63
    
    elif row['Currency'] == 'BRL':
        return row['Amount']*2.25
    
    elif row['Currency'] == 'BGN':
        return row['Amount']*6.23
    else:
        return row['Amount'] 

df['Amount (GHS)'] = df.apply(convert_to_ghs, axis=1)

# Print the DataFrame with converted amounts
print(df)
```
